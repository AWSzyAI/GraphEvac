```python
import pandas as pd
df = pd.read_csv("../output/select_result.csv")
df.shape
```

```python
df["makespan_td"] = pd.to_timedelta(df["makespan_hms"])
```

```python
best_each_floor = (
    df.loc[df.groupby("floors")["makespan_td"].idxmin()]
    [["floors", "responders", "exit_combo_id", "makespan_hms"]]
    .sort_values("floors")
)

best_each_floor
```

```python
for f in sorted(df["floors"].unique()):
    sub = df[df["floors"] == f]
    row = sub.loc[sub["makespan_td"].idxmin()]
    print(f"Floors={f:2d}, Best Responders={row.responders}, "
          f"ExitCombo={row.exit_combo_id}, Time={row.makespan_hms}")
```

```python
responder_summary = (
    df.groupby("responders")["makespan_td"]
    .mean()
    .apply(lambda td: str(td).split()[-1])      # 转回 HH:MM:SS
)

responder_summary
```

```python
best_by_responder = (
    df.loc[df.groupby("responders")["makespan_td"].idxmin()]
    [["responders","floors","exit_combo_id","makespan_hms"]]
    .sort_values("responders")
)
best_by_responder
```

```python
exit_summary = (
    df.groupby("exit_combo_id")["makespan_td"]
    .mean()
    .apply(lambda td: str(td).split()[-1])
)

exit_summary
```
